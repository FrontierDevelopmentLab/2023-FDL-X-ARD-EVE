```python
import boto3
from botocore import UNSIGNED
from botocore.config import Config
s3 = boto3.client('s3', config=Config(signature_version=UNSIGNED))
```

```python
# Specify the bucket name
bucket_name = 'gov-nasa-hdrl-data1'

# Initialize the object count
object_count = 0

# Paginate through the objects in the bucket
paginator = s3.get_paginator('list_objects_v2')
response_iterator = paginator.paginate(Bucket=bucket_name)

for response in response_iterator:
    if 'Contents' in response:
        object_count += len(response['Contents'])

# Display the number of objects in the bucket
print(f"Number of objects in the bucket: {object_count}")
```

```python
# Create an S3 client
s3_client = boto3.client('s3', config=Config(signature_version=UNSIGNED))

# Specify the bucket name
bucket_name_mi = 'gov-nasa-hdrl-data1'

# Call the list_objects_v2 API to retrieve object information
response_mi = s3_client.list_objects_v2(Bucket=bucket_name_mi)

# Get the count of objects in the bucket
object_count = response_mi['KeyCount']

# Display the number of objects in the bucket
print(f"Number of objects in the bucket: {object_count}")
```
