```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv("/Users/richardgalvez/data/fdl2023/virtual_eve_2010_2020.csv")
data['timestamp'] = pd.to_datetime(data['timestamp'])
data.set_index("timestamp", inplace=True)
data.sort_index(inplace=True)
data.drop(columns=["inference_time", "model_name"], inplace=True)
```

```python
data_subsample = data.resample("24H").mean()
```

```python
def remove_outliers(data, threshold=3):
    z_scores = np.abs((data - data.mean()) / data.std())
    data = data[(z_scores < threshold).all(axis=1)]
    return data

# Apply remove_outliers function to all columns except timestamp
data = data.apply(remove_outliers, axis=0)

# Plot all columns
data.plot(figsize=(10,5))
plt.show()
```
